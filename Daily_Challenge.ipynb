```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank'],
    'Age': [5, 16, 25, 45, 70, 12],
    'Score': [85, 90, 78, 92, 88, 75]
}

df = pd.DataFrame(data)

df.loc[2, 'Score'] = np.nan

mean_score = df['Score'].mean()
df['Score'] = df['Score'].fillna(mean_score)

bins = [0, 12, 18, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)

df_encoded = pd.get_dummies(df, columns=['Age_Group'], prefix='Group')

try:
    df_encoded.to_csv('final_dataset.csv', index=False)
    df_encoded.to_json('final_dataset.json', orient='records')
    print("✅ Exportation réussie : final_dataset.csv et final_dataset.json")
except Exception as e:
    print(f"❌ Erreur lors de l'exportation : {e}")

print("\n--- Résumé statistique ---")
display(df.describe())

print("\n--- Aperçu du dataset encodé ---")
display(df_encoded.head())
```
