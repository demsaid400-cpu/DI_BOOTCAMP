```python
import pandas as pd

sleep = pd.read_csv('Time Americans Spend Sleeping 2.csv')
mental = pd.read_csv('clean_dataset.csv')
credit = pd.read_csv('crx.csv')

def classify_column(col):
    return "Qualitative" if col.dtype == "object" else "Quantitative"

def get_summary(df, name):
    return pd.DataFrame({
        "Dataset": name,
        "Column": df.columns,
        "Type pandas": df.dtypes.values,
        "Classification": [classify_column(df[col]) for col in df.columns]
    })

full_summary = pd.concat([
    get_summary(sleep, "Sleep"),
    get_summary(mental, "Mental Health"),
    get_summary(credit, "Credit")
], ignore_index=True)

print("Analyse terminée pour les trois datasets.")
display(full_summary)
```
