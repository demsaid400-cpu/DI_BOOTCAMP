```python
import numpy as np
from scipy import stats

np.random.seed(42)

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_statistic, p_value = stats.ttest_ind(data1, data2)

print("Statistique T :", t_statistic)
print("P-value :", p_value)
alpha = 0.05

if p_value < alpha:
    print("Rejet de l'hypothèse nulle (H0)")
    print("Les moyennes des deux groupes sont significativement différentes.")
else:
    print("Impossible de rejeter H0")
    print("Aucune différence significative détectée.")
print("Moyenne Groupe 1 :", np.mean(data1))
print("Moyenne Groupe 2 :", np.mean(data2))
import matplotlib.pyplot as plt

plt.hist(data1, bins=15, alpha=0.7, label='Groupe 1')
plt.hist(data2, bins=15, alpha=0.7, label='Groupe 2')

plt.title("Comparaison des distributions")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.legend()

plt.show()
```

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
from scipy import stats

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F-value :", f_value)
print("P-value :", p_value)
import numpy as np

print("Moyenne Fertilizer 1 :", np.mean(fertilizer_1))
print("Moyenne Fertilizer 2 :", np.mean(fertilizer_2))
print("Moyenne Fertilizer 3 :", np.mean(fertilizer_3))
```

On observe clairement :

* Fertilizer 2 → meilleure croissance
* Fertilizer 3 → plus faible croissance

```python
import matplotlib.pyplot as plt

data = [fertilizer_1, fertilizer_2, fertilizer_3]

plt.boxplot(data, tick_labels=["F1", "F2", "F3"])
plt.title("Comparaison des engrais (ANOVA)")
plt.ylabel("Croissance des plantes (cm)")
plt.show()
```

Interprétation des résultats

1. Résultat du test

* F-value élevée → forte différence entre les groupes
* P-value très petite (< 0.05) → résultat statistiquement significatif

👉 Conclusion :

✔ Les engrais ont des effets significativement différents sur la croissance des plantes.

⸻

2. Que signifie une p-value > 0.05 ?

Si la p-value était supérieure à 0.05, cela signifierait :

* On ne rejette pas H₀
* Les différences observées entre les groupes pourraient être dues au hasard
* On conclurait que les engrais ont des effets similaires

```python
import pandas as pd

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})
from scipy.stats import pearsonr, spearmanr

age = data['age']
income = data['income']

pearson_corr, pearson_p = pearsonr(age, income)

spearman_corr, spearman_p = spearmanr(age, income)

print("Corrélation de Pearson :", pearson_corr)
print("P-value Pearson :", pearson_p)

print("Corrélation de Spearman :", spearman_corr)
print("P-value Spearman :", spearman_p)
```

Interprétation

1. Corrélation de Pearson

* ≈ 1.0
* Indique une relation linéaire positive presque parfaite

👉 Plus l’âge augmente, plus le revenu augmente de façon proportionnelle.

⸻

2. Corrélation de Spearman

* = 1.0
* Indique une relation monotone parfaite

👉 Le classement des âges correspond exactement au classement des revenus.

```python
import matplotlib.pyplot as plt

plt.scatter(data['age'], data['income'])

plt.title("Relation âge vs revenu")
plt.xlabel("Âge")
plt.ylabel("Revenu")
plt.show()
```

Résultat final :

* Pearson ≈ 1.0
* Spearman = 1.0

👉 Forte corrélation positive entre âge et revenu
